```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from numpy.linalg import inv
import random

N = 1000

std = 0.6

height = np.random.normal(175, std, N)
weight = np.random.normal(85, std, N)

height = height.reshape(N,1)
weight = weight.reshape(N,1)
```

```python
plt.scatter(height, weight)
```

```python
X = height
Y = weight
color = ['blue', 'yellow', 'orange', 'black', 'purple']
for i in range(5):
#     m = np.random.random(1)
    m = np.array([np.random.uniform(-0.4, 0.5)])
    Y_pred = np.matmul(X, m)
    plt.plot(height, Y_pred, c=color[i], lw = 4)
plt.scatter(height, weight, c='green')
```

```python
X = height
Y = weight
m = np.matmul(inv(np.matmul(X.T, X)), np.matmul(X.T, Y))
Y_pred = np.matmul(X, m)
plt.scatter(height, weight, c='green')
plt.plot(height, Y_pred, c='red', lw = 4)
```
