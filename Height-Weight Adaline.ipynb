```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from numpy.linalg import inv
import random

N = 200

std = 10

# height = np.random.normal(175, std, N)
# weight = np.random.normal(85, std, N)

height = np.random.random_sample((N,1))*100 + 40
e = np.random.rand(N) * std
weight = np.matmul(height, [0.3]).T + e
```

```python
plt.scatter(height, weight)
```

```python
height = height.reshape(N,1)
weight = weight.reshape(N,1)
X = height
Y = weight
color = ['blue', 'yellow', 'orange', 'black', 'purple']
for i in range(5):
    m = np.array([np.random.uniform(0.2, 0.5)])
    Y_pred = np.matmul(X, m)
    plt.plot(height, Y_pred, c=color[i], lw = 4)
plt.scatter(height, weight, c='green')
```

```python
X = height
Y = weight
m = np.matmul(inv(np.matmul(X.T, X)), np.matmul(X.T, Y))
Y_pred = np.matmul(X, m)
plt.scatter(height, weight, c='green')
plt.plot(height, Y_pred, c='red', lw = 4)
```

```python
import scipy
result = scipy.stats.linregress(X[:,0], Y[:,0])
Y_pred = result.intercept + np.matmul(X, np.array([result.slope]))
plt.scatter(height, weight, c='green')
plt.plot(height, Y_pred, c='red', lw = 4)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
clf=LinearRegression(fit_intercept=False)
clf.fit(X, Y)
y_pred=clf.predict(X)
```

```python
plt.scatter(height, weight, c='green')
plt.plot(height, Y_pred, c='red', lw = 4)
```
