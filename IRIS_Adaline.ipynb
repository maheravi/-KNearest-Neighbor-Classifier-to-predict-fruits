```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from numpy.linalg import inv

N = 100

std = 1

hair_strand_length_man = np.random.normal(5, std, N)
hair_strand_length_woman = np.random.normal(20, std, N)

man_label = np.zeros(N, dtype='int')
woman_label = np.ones(N, dtype='int')

# X = np.array([hair_strand_length_man, hair_strand_length_woman]).T
X = np.concatenate((hair_strand_length_man, hair_strand_length_woman))
Y = np.concatenate((man_label, woman_label))
```

```python
plt.scatter(X[0:N], Y[0:N], c='orange', label='Man')
plt.scatter(X[N:2*N], Y[N:2*N], c='blue', label='Woman')

plt.xlabel('strand length')
plt.ylabel('Sex')
plt.title("Man and Woman Hair Strand Length")
plt.legend()
```

```python
plt.scatter(X[0:N], Y[0:N], c='orange', label='Man')
plt.scatter(X[N:2*N], Y[N:2*N], c='blue', label='Woman')

plt.xlabel('strand length')
plt.ylabel('strand length')
plt.title("Man and Woman Hair Strand Length")
plt.legend()
```

```python
Y = Y.reshape(200,1)
X = X.reshape(200,1)

m = np.matmul(inv(np.matmul(X.T, X)), np.matmul(X.T, Y))
Y_pred = np.matmul(X, m)
plt.scatter(X[0:N], Y[0:N], c='orange', label='Man')
plt.scatter(X[N:2*N], Y[N:2*N], c='blue', label='Woman')
plt.plot(X, Y_pred, c='red', lw = 2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```
